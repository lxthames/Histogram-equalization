```python
#Backend Coding of Histogram
%matplotlib inline
from IPython.display import display, Math, Latex

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image


#Conversion into Greyscale in PLT


img = Image.open('image(1).jpg').convert('L')

plt.imshow(img, cmap='gray')
```

```python
img = np.asarray(img)
flat = img.flatten()
plt.hist(flat,bins=100)  #bins=50 is the number of values in array, distribution 
#this first step is the graph of frequencies
```

```python
def get_histo(image,bins):
  #array with size of bins, set to zeros
  histogram=np.zeros(bins)

  #loop through pixels and sum up counts of pixels
  for pixel in image:
    histogram[pixel]+=1
  
  #return our final result
  #plt.plot(histogram)
  return histogram


#execute our histogram function
hist = get_histo(flat,256)

#create our cumulative sum function, graph increasing because sum adding and adding
def cumsum(a):
  a=iter(a)
  b= [next(a)]
  for i in a:
    b.append(b[-1]+i)
  return np.array(b)
  #execute the fn

cs =cumsum(hist)
plt.plot(cs)
```

```python
#numerator and denomenator
nj = (cs.min())*255
N=cs.max()-cs.min()
```
